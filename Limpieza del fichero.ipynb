```python
# aqui me hice una funcion para limpiarme todo el money raised. Voy a copiar la de Noelia porque consegui hacerlo
# a mi manera. He perdido todo lo que tenia del fin de semana y no tengo tiempo de volverme a pensar la locular
# de regex que hice. Obviamente este comentario cuando suba mi repositorio github en algun sitio lo borro.
import re

def moneyRaise(value):
    dicc_coin = {'CAD': 1.32,'RUB': 63.75, 'EUR': 0.90, 'GBP': 0.77}
    values_money = {'K':1000, 'M':1000000, 'B': 100000000000}
    value_number = float(re.search('[+-]?([0-9]*[.])?[0-9]+', value)[0])
    if value.endswith('B'):
        cambio = value_number*(values_money['B'])
    elif value.endswith('K[k]'):
        cambio = value_number*(values_money['K'])
    elif value.endswith('M'):
        cambio = value_number*(values_money['M'])
    elif value.startswith("C"):
        cambio =  value_number*(dicc_coin['CAD'])
    elif value.startswith("$"):
        cambio =  value_number
    elif value.startswith("€"):
        cambio = value_number*(dicc_coin['EUR'])
    elif value.startswith("£"):
        cambio = value_number*(dicc_coin['GBP'])
    elif value.startswith("r"):
        cambio = value_number*(dicc_coin['RUB'])
    else:
        cambio = value_number
    return int(cambio)
```

```python
df4['total_money_raised'] = df4['total_money_raised'].apply(moneyRaise)
```
